**1. EV Adoption Trends**

**Analyzing the adoption rate of electric vehicles over the years.**

```python
# Calculate the yearly totals
yearly_totals=evsales.iloc[:,2:].sum()
print("Yearly EV Production Totals:\n",yearly_totals)
```

**How has the demand for EVs changed over time?**

```python
# Calculate Year-over-Year (YoY) growth in EV production
yoy_growth=yearly_totals.pct_change()*100
print("\nYear-over-Year Growth(%):\n",yoy_growth)

# Determine the overall trend by comparing first and last year
initial_value = yearly_totals.iloc[0]
final_value = yearly_totals.iloc[-1]
overall_growth=((final_value-initial_value)/initial_value)*100
print(f"\nOverall Growth from {yearly_totals.index[0]} to {yearly_totals.index[-1]}:{overall_growth:.2f}%")
```

**2. Manufacturer Performance**

**Compare the performance of different EV manufacturers.**

```python
#First, we need to create a 'Total Sales' column by summing sales across the years
evsales['Total Sales']=evsales.iloc[:,2:].sum(axis=1)

#Grouping data by manufacturer(Maker) to calculate total sales
manufacturer_sales=evsales.groupby('Maker')['Total Sales'].sum().reset_index()

#Sorting manufacturers by total sales in descending order
manufacturer_sales=manufacturer_sales.sort_values(by='Total Sales',ascending=False)
print(manufacturer_sales.head(10))
```

**Which companies have the highest market share, and what factors contribute to their success?**

```python
# Calculate total sales across all manufacturers
total_market_sales=manufacturer_sales['Total Sales'].sum()
print("total_market_sales:\n",total_market_sales)

# Calculate market share for each manufacturer
manufacturer_sales['Market Share(%)']=(manufacturer_sales['Total Sales']/total_market_sales)*100
manufacturer_sales.head(10)
```

**5. Geographical Analysis**

**Study how the demand for electric vehicles varies across different regions in India.Which regions are leading in EV adoption, and what might be the reasons behind this trend?**

```python
# Counting number of EV manufacturers by state
ev_makers_by_state=evmarketplace.groupby('State')['EV Maker'].count().reset_index()
ev_makers_by_state.columns=['State','Number of EV Makers']

# Displaying EV manufacturers count per state
ev_makers_by_state.head()
# Merging the number of EV makers and operational charging stations per state
state_ev_analysis=pd.merge(ev_makers_by_state,operationalpc,on='State',how='left')

# Displaying the merged analysis with EV makers and charging stations per state
state_ev_analysis.head()
# Sorting states by number of EV makers and operational charging stations
top_ev_states=state_ev_analysis.sort_values(by=['Number of EV Makers','No. of Operational PCS'],ascending=False)

# Displaying the top states with the most EV infrastructure
top_ev_states.head()
```

```python
#Grouping data by state to get the count of EV manufacturers and the total number of charging stations
ev_maker_count_by_state=evmarketplace.groupby('State').size().reset_index(name='No. of EV Manufacturers')

#Merge EV manufacturer count with operational charging stations data
geographical_ev_analysis=operationalpc.merge(ev_maker_count_by_state,on='State',how='left').fillna(0)

#Sort by the number of operational charging stations and EV manufacturers to identify leading regions
geographical_ev_analysis=geographical_ev_analysis.sort_values(by=['No. of Operational PCS','No. of EV Manufacturers'],ascending=False)

#Display the results
print(geographical_ev_analysis)
```
